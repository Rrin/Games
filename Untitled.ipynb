## Q.1 Parametric Functions

```python
def curve(A,B, C):
    """
    this function plots the parametric curve 
    given integers A, B and C
    """
    twopi = float(2*(22/7))
    # create a vector of 100 numbers between 0, and 2pi
    t = np.linspace(0,twopi,100)
    sin = np.vectorize(math.sin)
    cos = np.vectorize(math.cos)
    #calculate x and y values
    x = A * (sin(t)**3)
    y = (B * cos(t)) - (C * (cos(2*t)))
    #plot x and y
    plt.axis('equal')
    plt.plot(x, y)
    plt.show()
    
```

```python
curve(13, 6, 4)
```

## Q.2 Power Series

```python
def power_series(a, x):
    """
    This function takes in a numpy vector a,
    and integer x and does the following:
    for each element of a , 
    multiply x to the power of the index of a
    """
    #number of elements of 1 dimesional array. Size is equal to N
    size = a.size
    # create a vector of numbers representing k increasing upto N
    k = np.arange(0, size)
    # calculate x*k for all N
    h = x**k
    #do a dot product of a and h
    hXa= np.dot(a,h)
    # do a cumulative sum over all numbers of h X a
    return np.sum(hXa)
    
```

```python
power_series(np.ones(100),0.5)
```

```python
from scipy.special import factorial
power_series(1 / factorial(np.arange(0,100)), 1)
```

## Q.3 Fourier Series

$$f_{triangle }\left( t\right) =\dfrac {8} {\pi ^{2}}\sum _{k=0}^{\infty }\left( -1\right) ^{k}\dfrac {sin\left( 2\pi \left( 2k+1\right) t\right) } {\left( 2k+1\right) ^{2}}$$

```python
def triangle_wave(N, interval):
    """
    This function plots a triangular wave via a fourier seiries
    for N cycles and interval specified by a list
    """
    y = []
    x = np.linspace(interval[0],interval[1],100)
    for t in np.linspace(0,N,100):
        pi = float(22/7)
        k = np.arange(0,N)
        sin = np.vectorize(math.sin)
        numerator  = sin((2*pi*t) * ((2*k) + 1))
        denominator = ((2*k) + 1)**2
        prod = (-1)**k
        #sum over all k's upto N
        summation = np.sum(prod * (np.divide(numerator,denominator)))
        fourier = (8 / (pi**2)) * summation
        y.append(fourier)
    plt.plot(x, y)
    plt.grid()
    #set y axis height
    axes = plt.gca()
    axes.set_ylim([-N-1,N-1])
    plt.show()
    
```

```python
triangle_wave(3, [0,3])
```

## Q.4 Fourier series: sawtooth wave

$$f_{sawtooth}\left( t\right) =\dfrac {2} {\pi }\sum _{k=1}^{\infty }\dfrac {\left( -1\right) ^{k+1}\sin \left( 2\pi kt\right) } {k}$$

```python
def sawtooth_waves(n, m, T):
    """
    fourier series: sawtooth for different values of N
    """
    fig = plt.figure(figsize=(15, 4))
    #loop through all graphs
    for sd in range(n*m):
        x = np.linspace(0,T,100)
        y = []
        for t in x:
            k = np.arange(1,sd+2)
            pi = float(22/7)
            sin = np.vectorize(math.sin)
            numerator = ((-1)**(k+1))* sin(2*pi*t*k)
            denominator = k
            # cumulative sum over all k's uptoN
            summation = np.sum(np.nan_to_num(numerator/denominator))
            fourier = (2 / pi) * summation
            y.append(fourier)
        #plotting a sub plot to each figure
        sub1 = fig.add_subplot(240+sd+1) # equivalent to: plt.subplot(2, 2, 1)
        plt.title('N = '+str(sd+1))
        sub1.grid()
        sub1.plot(x, y)
    fig.tight_layout()        
    
```

```python
sawtooth_waves(2, 4, 2)
```

## Q.5 Riemann zeta function

$$\varsigma=\sum _{n=1}^{\infty }\dfrac {1} {n_{s}}=\prod _{p prime}\dfrac {1} {1-s^{-s}}=\dfrac {1} {1-2^{-s}}\cdot \dfrac {1} {1-3^{-s}}\cdot\dfrac {1} {1-3^{-s}}\cdot\dfrac {1} {1-7^{-s}}\ldots\dfrac {1} {1-p^{-s}}\ldots$$

```python
import functools
def primes(n):
    """ 
    generate and
    return  a list of primes  for < = n 
    """
    n = n+1
    sc = np.arange(3, n, 2)
    for nos in range(3, int(n**0.5) + 1, 2): 
        if sc[(nos-3) / 2]: 
            sc[(nos*nos-3) / 2::nos]=0
    return np.r_[2, sc[sc>0]]

def euler_product(s, N):
    """
    This function computes a euler product using 
    the Riemann zeta function
    """
    primeNos = primes(N)
    denom = 1 - (primeNos ** -float(s))
    result = 1 / denom
    #reduce product
    cumulatiVeProduct = functools.reduce(lambda x, y: x*y, result)
    return cumulatiVeProduct
    
```

```python
euler_product(4, 100000)
```

## Q.6 slope field for diffrential equations

```python
def f(t,y):
    return y**2 - t**2

def slope_field(f, tlims, ylims, grid_step):
    """
    This function will take in a diffrential equation x and 
    render its slope field while being bounded by grid limits
    tlims and ylims
    """
    x = np.linspace(tlims[0],tlims[1],20)
    y = np.linspace(ylims[0],ylims[1],20)
    #loop though getting midpoints j,k
    for j in x:
        for k in y:
            #calculate y'
            slope = f(j,k)
            #crete an array for midp0iints
            domain = np.linspace(j-grid_step/2,j+grid_step/2,2)
            def fd(x1a,y1a):
                za = slope*(domain-x1a)+y1a
                return za
            plt.plot(domain, fd(j,k), color = '#340B8C')
    axes = plt.gca()
    axes.set_xlim(tlims)
    axes.set_ylim(ylims)
    plt.grid(True)
    plt.show()
```
